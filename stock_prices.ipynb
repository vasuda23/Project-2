```python
response_amzn = requests.request("GET", url, headers=headers, params=querystring_amzn)
```

```python
amzn_stockinfo = response_amzn.json()

amzn_info = amzn_stockinfo['prices']

amzn_introdf = pd.DataFrame(amzn_info)
amazon = amzn_introdf.drop(['adjclose'], axis=1)
amazon['date'] = pd.to_datetime(amazon['date'], unit='s')
amazon['date'] = pd.to_datetime(amazon['date']).dt.date
amazon
```

```python
querystring_jwn = {"frequency":"1d","filter":"history","period1":"1577984400","period2":"1593709200","symbol":"JWN"}
response_jwn = requests.request("GET", url, headers=headers, params=querystring_jwn)
```

```python
querystring_mfa = {"frequency":"1d","filter":"history","period1":"1577984400","period2":"1593709200","symbol":"MFA"}
response_mfa = requests.request("GET", url, headers=headers, params=querystring_mfa)
```
